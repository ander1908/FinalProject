```python
frames = [ts20filt_df,ts21filt_df]
tstatstotal_df = pd.concat(frames)
tstatstotal_df.head()
```

```python
tst_df = tstatstotal_df
```

```python
tst_df["Goals"].count()
```

```python
y=tst_df["Stanley Cup"]
winner_losers = ["negative","positive"]
```

```python
X=tst_df.drop("Stanley Cup", axis=1)
#need to drop team name as well
X=X.drop("Name",axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.5)
```

```python
#len(X_train)
len(y_train)
```

```python
from sklearn.svm import SVC 
```

```python
#linear fit is not great, lets try polynomial
#poly is slightly better at model = SVC(kernel='poly', degree=8)
#gaussian is terribel
model = SVC(kernel='poly', degree=8)
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=winner_losers))
```
